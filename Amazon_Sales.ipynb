### Data type

Let's take a look at the data types and check if there are any columns with inappropriate data types that can be processed further.

```python
df.dtypes
```

It can be seen that these columns: "discounted_price", "actual_price", "discount_percentage", "rating", "rating_count" should have numeric data type instead of object. So I will handle this by changing Data Types of Columns from **object** to **float**

```python
# Changing the data type of discounted price 

df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')
```

```python
# Changing the data type of actual price

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
# Changing Datatype and values in Discount Percentage

df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')
df['discount_percentage'] = df['discount_percentage'] / 100
```

```python
# Changing the data type of rating

df['rating'] = df['rating'].str.replace(r'[^0-9.]', '', regex=True)
#df['rating'] = df['rating'].astype('float64')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')  # 'coerce' converts errors to NaN
```

```python
# Changing the data type of rating count

df['rating_count'] = df['rating_count'].str.replace(",",'')
df['rating_count'] = df['rating_count'].astype('float64')
```

### Data distribution

#### For each column with numeric data type, how are the values distributed?

For columns with numeric data types, we calculate:

- Percentage (from 0 to 1) of missing values
- Min value
- Lower quartile value (25th percentile)
- Median value (50th percentile)
- Upper quartile value (75th percentile)
- Max value

The group will save the results to the DataFrame `num_col_info_df`, where:
- Column names are the names of the numeric columns in `df`
- Row names: “missing_ratio”, “min”, “low_quartile”, “median”, “upper_quartile”, “max”

For ease of viewing, the group rounds all values to 1 decimal place using the `.round(1)` method.

```python
num_col_info_df = None


def missing_values_count(series):
    return series.isna().sum()
    
# Define missing ratio    
def missing_ratio(series):
    percentage = series.isna().sum() / len(series)
    return percentage.round(1)

# Define lower quartile
def lower_quartile(series):
    return series.quantile(0.25).round(1)

# Define upper quartile
def upper_quartile(series):
    return series.quantile(0.75).round(1)

#Define median
def median(series):
    return series.quantile(0.75).round(1)

num_col_info_df = df.select_dtypes(include=['int64', 'float64'])

num_col_info_df = num_col_info_df.agg([missing_values_count, missing_ratio, "min", lower_quartile, median, upper_quartile, "max"])

num_col_info_df
```

##### Comment

- The `missing_ratio` row shows 0.0 for all columns, indicating there are almost no missing data (only 1 missing value in `rating` and 2 missing values in `rating_count`).
- The data shows right-skewed distributions in `discounted_price`, `actual_price`, and `rating_count`, indicating a few expensive products and highly rated/popular items.
- The `discount_percentage` suggests most products have moderate to high discounts (30%-60%).
- Ratings are generally high, with most products rated above 4.0.

- Since the number of missing values are so small, let's check the issue with the missing rows.

```python
df[df['rating_count'].isna()].head(5)
```

```python
df[df['rating'].isna()].head(5)
```

- As we can see, the missing values are not in the same rows so I will handle each column.
    - With the missing value in `rating`, I have  searched the provided product_id on their official website (amazon.in) and got the rating is 3.9. So, I am going to give the item rating a 3.9 as well.
    - With the missing values in `rating_count`, I decided to impute by median because it works well for skewed data and preserves the overall distribution.

```python
# Impute missing values

df['rating'] = df.rating.fillna(value=3.9)
df['rating_count'] = df.rating_count.fillna(value=df['rating_count'].median())
```

Let's check our data again

```python
num_col_info_df = df.select_dtypes(include=['int64', 'float64'])

num_col_info_df = num_col_info_df.agg([missing_values_count, missing_ratio, "min", lower_quartile, median, upper_quartile, "max"])

num_col_info_df
```

#### For each column with categorical data type, how are the values distributed?

For columns with the data type `categorical`, we need to calculate:

- The percentage (from 0 to 1) of missing values
- The number of different values (not considering missing values)
- The percentage (from 0 to 100) of each value sorted by decreasing percentage (not considering missing values)

We will save the results to the DataFrame `cat_col_info_df`, where:
- Column names are the non-numeric column names in `df`
- Row names are: “missing_ratio”, “num_values”, “value_ratios”

For ease of viewing, we round all values to 1 decimal place using the `.round(1)` method.

```python
pd.set_option('display.max_colwidth', 100) # For clearly
pd.set_option('display.max_columns', None) # For clearly
```

```python
cat_col_info_df = df.select_dtypes(include='object')

# Define num values
def num_values(series):
    return len(series.dropna().unique())

# Define value ratios
def value_ratios(series):
    value_counts = series.dropna().value_counts()
    percentages = (value_counts / len(series))*100
    percentages = percentages.round(1)
    sorted_percentages = percentages.sort_values(ascending=False)
    return sorted_percentages.to_dict()
cat_col_info_df = cat_col_info_df.agg([missing_values_count, missing_ratio, num_values, value_ratios])
cat_col_info_df
```

```python
from textblob import TextBlob

# Select relevant columns for sentiment analysis
temp_df = processed_df[["product_id", "user_id", "review_content"]].copy()

# Calculate sentiment polarity score for each review using TextBlob
temp_df["sentiment_score"] = temp_df["review_content"].dropna().apply(lambda text: TextBlob(str(text)).sentiment.polarity)

# Define sentiment categories based on score
def classify_sentiment(score):
    if score > 0.05:
        return "Positive"
    elif score < -0.05:
        return "Negative"
    else:
        return "Neutral"

# Apply classification
temp_df["sentiment_label"] = temp_df["sentiment_score"].apply(classify_sentiment)

# Count sentiment distribution
sentiment_counts_textblob = temp_df["sentiment_label"].value_counts()

# Display top 10 most positive reviews
top_positive_reviews = temp_df.sort_values(by="sentiment_score", ascending=False).head(10)

sentiment_counts_textblob
```
